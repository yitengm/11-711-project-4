```python
import pandas as pd
import concurrent.futures
from googletrans import Translator
import time

# Load your DataFrame
# Adjust the path for training dataset, testing dataset, validation dataset
train_data = pd.read_csv("/content/drive/MyDrive/11711/test_split.csv")

def translate_text(text, dest_language='en'):
    translator = Translator()
    # Check if the text is a string and not empty
    if isinstance(text, str) and text.strip():
        try:
            translation = translator.translate(text, dest=dest_language)
            return translation.text
        except Exception as e:
            print(f"Error translating text: {e}")
            print(text)
            return text  # Return the original text in case of an error
    else:
        # Return text as is if it's not a valid string
        return text



# Translate each row individually
for index, row in train_data.iterrows():
    if index < 16200:
      continue
    translated_text = translate_text(row['Utterance'])
    # print(translated_text)
    train_data.at[index, 'translation'] = translated_text if translated_text is not None else row['Utterance']

    # Save the file every 1000 rows
    if index % 200 == 0:
        train_data.to_csv("/content/drive/MyDrive/11711/test_split_translated_full.csv", index=False) # change the output path if needed
        print(f"Translation completed up to index {index} and saved to new CSV file.")

    time.sleep(0.3)

    if index % 600 == 0:
        time.sleep(60)

# Save the final version of the DataFrame
# modify the path if needed
train_data.to_csv("/content/drive/MyDrive/11711/test_split_translated_full_v2.csv", index=False)
```
